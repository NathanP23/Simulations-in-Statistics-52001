## Overview

`Part2l(Lambda, lambda, I)` returns the points of a non-homogeneous Poisson process restricted to an interval $I$. The rate function on $(0, \infty)$ is proportional to $e^{-\lambda x}$, so its expected total count is $\Lambda$.

**Construction (rejection / restriction):**

1. Draw $N \sim \operatorname{Poisson}(\Lambda)$.
2. If $N = 0$, return an empty numeric vector.
3. Draw $X_1, \ldots, X_N \stackrel{\text{iid}}{\sim} \operatorname{Exp}(\lambda)$ (the density $\propto e^{-\lambda x}$ on $(0,\infty)$ is $\operatorname{Exp}(\lambda)$).
4. Restrict to $I = (I_1, I_2)$: keep only $X_i \in I$. If none survive, return an empty vector.

A subset of a Poisson process is still a Poisson process, so the surviving points form an NHPP on $I$.

## R Implementation

```r
Part2l <- function(Lambda, lambda, I) {
  poisson_total <- rpois(1, Lambda)
  if (poisson_total == 0) return(numeric(0))
  candidate_x   <- rexp(poisson_total, rate = lambda)
  inside_I      <- candidate_x[candidate_x >= I[1] & candidate_x <= I[2]]
  return(inside_I)
}
```

```r
set.seed(1)

total_intensity   <- 19.7
rate_lambda       <- 1.28
interval_I        <- c(0.53, 1.36)
split_point_b     <- 0.95
n_replicates      <- 1e5

cat("Lambda    =", total_intensity, "\n")
cat("lambda    =", rate_lambda,     "\n")
cat("I         =", interval_I,      "\n")
cat("b         =", split_point_b,   "\n")
cat("reps      =", n_replicates,    "\n")
```

## Computing the Results

```r
test.function <- function(X, b) {
  N1 <- sum(X <= b)
  N2 <- sum(X >  b)
  return(c(N1, N2))
}

temp <- replicate(
  n_replicates,
  test.function(
    Part2l(Lambda = total_intensity, lambda = rate_lambda, I = interval_I),
    b = split_point_b
  )
)
N1 <- temp[1, ]
N2 <- temp[2, ]

freq_N1_eq_4 <- mean(N1 == 4)
freq_N2_eq_2 <- mean(N2 == 2)
correlation  <- cor(N1, N2)

cat("dim(temp)        =", dim(temp), "\n")
cat("a. mean(N1 == 4) =", freq_N1_eq_4, "\n")
cat("b. mean(N2 == 2) =", freq_N2_eq_2, "\n")
cat("c. cor(N1, N2)   =", correlation,  "\n")
```
