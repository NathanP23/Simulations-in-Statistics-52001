## Overview

`Part2g(n, pi.0, P)` returns $X_1, \ldots, X_n$ from a finite-state Markov chain on the support $\{1, 2, \ldots, k\}$:

$$X_0 \sim \pi_0, \qquad X_i \mid X_{i-1} \sim P[X_{i-1}, \cdot],\quad 1 \le i \le n$$

where $\pi_0$ is a length-$k$ probability vector for the initial state and $P$ is a $k \times k$ row-stochastic transition matrix.

## R Implementation

```r
Part2g <- function(n, pi.0, P) {
  k          <- length(pi.0)
  cum_rows_P <- t(apply(P, 1, cumsum))
  uniforms   <- runif(n)
  observations <- integer(n)
  current_state <- sample.int(k, size = 1, prob = pi.0)
  for (step_index in seq_len(n)) {
    current_state <- findInterval(uniforms[step_index], cum_rows_P[current_state, ]) + 1L
    observations[step_index] <- current_state
  }
  return(observations)
}
```

```r
set.seed(1)

n_steps        <- 1e6
initial_dist   <- c(0.2592514, 0.347922, 0.3928266)
transition_mat <- matrix(c(
  0.09, 0.50, 0.41,
  0.17, 0.30, 0.53,
  0.45, 0.29, 0.26
), nrow = 3, ncol = 3, byrow = TRUE)

cat("n              =", n_steps, "\n")
cat("k (states)     =", length(initial_dist), "\n")
cat("pi.0:\n");          print(initial_dist)
cat("rowSums(pi.0)  =", sum(initial_dist), "\n")
cat("P:\n");             print(transition_mat)
cat("rowSums(P)     =", rowSums(transition_mat), "\n")
```

## Computing the Results

```r
X <- Part2g(n = n_steps, pi.0 = initial_dist, P = transition_mat)

freq_state_1 <- mean(X == 1)
freq_state_2 <- mean(X == 2)
freq_state_3 <- mean(X == 3)

cat("length(X)        =", length(X),    "\n")
cat("a. mean(X == 1)  =", freq_state_1, "\n")
cat("b. mean(X == 2)  =", freq_state_2, "\n")
cat("c. mean(X == 3)  =", freq_state_3, "\n")
```
