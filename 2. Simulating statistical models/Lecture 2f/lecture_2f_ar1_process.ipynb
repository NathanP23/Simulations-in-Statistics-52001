## Overview

`Part2f(n, a1, sigma, mu0, sigma0)` returns $X_1, \ldots, X_n$ from the AR(1) recursion

$$X_i = a_1 X_{i-1} + \varepsilon_i, \qquad \varepsilon_i \stackrel{\text{iid}}{\sim} \mathcal{N}(0, \sigma^2),\quad 1 \le i \le n$$

initialized with $X_0 \sim \mathcal{N}(\mu_0, \sigma_0^2)$. Defaults: $\mu_0 = 0$ and $\sigma_0 = \sigma/\sqrt{1 - a_1^2}$ (the stationary value when $|a_1| < 1$).

## R Implementation

```r
Part2f <- function(n, a1, sigma,
                   mu0    = 0,
                   sigma0 = sqrt(sigma^2 / (1 - a1^2))) {
  X0_init       <- rnorm(1, mean = mu0, sd = sigma0)
  innovations   <- rnorm(n, mean = 0,   sd = sigma)
  observations  <- as.numeric(filter(innovations,
                                     filter = a1,
                                     method = "recursive",
                                     init   = X0_init))
  return(observations)
}
```

```r
set.seed(1)

n_obs        <- 1e6
ar1_coef     <- -0.63
noise_sigma  <- 2.16

cat("n           =", n_obs,       "\n")
cat("a1          =", ar1_coef,    "\n")
cat("sigma       =", noise_sigma, "\n")
cat("mu0         = (default) 0\n")
cat("sigma0      = (default) sqrt(sigma^2/(1-a1^2)) =",
    sqrt(noise_sigma^2 / (1 - ar1_coef^2)), "\n")
```

## Computing the Results

```r
X <- Part2f(n = n_obs, a1 = ar1_coef, sigma = noise_sigma)

sd_X        <- sd(X)
lag1_cor    <- cor(X[-1], X[-length(X)])

cat("length(X)             =", length(X), "\n")
cat("a. sd(X)              =", sd_X,      "\n")
cat("b. cor(X[-1], X[-n])  =", lag1_cor,  "\n")
```
