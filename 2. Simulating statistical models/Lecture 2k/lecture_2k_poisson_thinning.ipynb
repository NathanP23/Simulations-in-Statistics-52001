## Overview

`Part2k(lambda, p)` returns a length-$2$ vector $(N_1, N_0)$ of two **independent** Poisson random variables built via the thinning principle:

$$N \sim \operatorname{Poisson}(\lambda), \qquad X_1, \ldots, X_N \stackrel{\text{iid}}{\sim} \operatorname{Bernoulli}(p_1)$$

$$N_1 = \sum_{i=1}^{N} X_i, \qquad N_0 = \sum_{i=1}^{N} (1 - X_i)$$

Then $N_1 \sim \operatorname{Poisson}(p_1 \lambda)$, $N_0 \sim \operatorname{Poisson}(p_2 \lambda)$ with $p_2 = 1 - p_1$, and $N_1 \perp N_0$. Equivalently, draw $N$ then split via a single $\operatorname{Binomial}(N, p_1)$.

## R Implementation

```r
Part2k <- function(lambda, p) {
  poisson_total   <- rpois(1, lambda)
  thinned_first   <- rbinom(1, size = poisson_total, prob = p[1])
  thinned_second  <- poisson_total - thinned_first
  c(thinned_first, thinned_second)
}
```

```r
set.seed(1)

poisson_lambda     <- 18.8
thinning_prob      <- c(0.46, 0.54)
n_replicates       <- 1e5

cat("lambda          =", poisson_lambda,    "\n")
cat("p               =", thinning_prob,     "\n")
cat("sum(p)          =", sum(thinning_prob),"\n")
cat("number of reps  =", n_replicates,      "\n")
```

## Computing the Results

```r
temp <- replicate(n_replicates, Part2k(lambda = poisson_lambda, p = thinning_prob))
N1   <- temp[1, ]
N2   <- temp[2, ]

freq_N1_eq_9  <- mean(N1 == 9)
freq_N2_eq_10 <- mean(N2 == 10)
correlation   <- cor(N1, N2)

cat("dim(temp)         =", dim(temp), "\n")
cat("a. mean(N1 == 9)  =", freq_N1_eq_9,  "\n")
cat("b. mean(N2 == 10) =", freq_N2_eq_10, "\n")
cat("c. cor(N1, N2)    =", correlation,   "\n")
```
