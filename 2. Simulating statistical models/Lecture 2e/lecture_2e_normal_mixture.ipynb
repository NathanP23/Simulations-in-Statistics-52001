## Overview

`Part2e(n, mu, sigma, theta)` returns a length-$n$ sample from the mixture density

$$f(x) = \sum_{a=1}^{k} \theta_a\, \phi(x \mid \mu_a, \sigma_a)$$

where $\phi(\cdot \mid \mu_a, \sigma_a)$ is the $\mathcal{N}(\mu_a, \sigma_a^2)$ density and $\theta = (\theta_1, \ldots, \theta_k)$ is a probability vector over the $k$ components.

**Sampling procedure:** for each of $n$ observations, draw a component index $a \in \{1, \ldots, k\}$ with probabilities $\theta$, then draw from $\mathcal{N}(\mu_a, \sigma_a^2)$.

## R Implementation

```r
Part2e <- function(n, mu, sigma, theta) {
  k                <- length(mu)
  component_index  <- sample.int(k, size = n, replace = TRUE, prob = theta)
  observations     <- rnorm(n, mean = mu[component_index], sd = sigma[component_index])
  return(observations)
}
```

```r
set.seed(1)

n_obs         <- 1e6
mixture_mu    <- c( -1.98, 0.22, 2.14)
mixture_sigma <- c( 3.48, 3.47, 4.01)
mixture_theta <- c( 0.27, 0.33, 0.4)

cat("n              =", n_obs,                 "\n")
cat("k (components) =", length(mixture_mu),    "\n")
cat("mu:\n");    print(mixture_mu)
cat("sigma:\n"); print(mixture_sigma)
cat("theta:\n"); print(mixture_theta)
cat("sum(theta)     =", sum(mixture_theta),    "\n")
```

## Computing the Results

```r
X <- Part2e(
  n     = n_obs,
  mu    = mixture_mu,
  sigma = mixture_sigma,
  theta = mixture_theta
)

mean_X <- mean(X)
sd_X   <- sd(X)

cat("length(X)  =", length(X), "\n")
cat("a. mean(X) =", mean_X,    "\n")
cat("b. sd(X)   =", sd_X,      "\n")
```

## Final Answer

With `set.seed(1)`:

$$\text{a.}\;\; \operatorname{mean}(X) \approx 0.5489 \qquad \text{b.}\;\; \operatorname{sd}(X) \approx 4.6676$$
