## Overview

`Part2i(P)` returns the stationary distribution $\pi_0$ of a row-stochastic transition matrix $P$, i.e. the probability vector satisfying

$$\pi_0\, P = \pi_0$$

Equivalently, $\pi_0^\top$ is a left eigenvector of $P$ with eigenvalue $1$, or a right eigenvector of $P^\top$ with eigenvalue $1$. We extract this eigenvector and normalize it to sum to $1$.

## R Implementation

```r
Part2i <- function(P) {
  eigen_decomp        <- eigen(t(P))
  leading_eigvec_real <- Re(eigen_decomp$vectors[, 1])
  stationary_dist     <- leading_eigvec_real / sum(leading_eigvec_real)
  return(stationary_dist)
}
```

```r
transition_mat <- matrix(c(
  0.15, 0.14, 0.71,
  0.77, 0.03, 0.20,
  0.11, 0.25, 0.64
), nrow = 3, ncol = 3, byrow = TRUE)

cat("k (states)     =", nrow(transition_mat),     "\n")
cat("P:\n");             print(transition_mat)
cat("rowSums(P)     =", rowSums(transition_mat),  "\n")
```

## Computing the Results

```r
pi.0 <- Part2i(transition_mat)

cat("Stationary distribution pi.0:\n")
cat("a. pi.0[1] =", pi.0[1], "\n")
cat("b. pi.0[2] =", pi.0[2], "\n")
cat("c. pi.0[3] =", pi.0[3], "\n")
cat("sum        =", sum(pi.0), "\n")
cat("check pi.0 %*% P:\n"); print(as.vector(pi.0 %*% transition_mat))
```
