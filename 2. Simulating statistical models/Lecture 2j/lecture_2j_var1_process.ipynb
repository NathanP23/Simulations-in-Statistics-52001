## Overview

`Part2j(n, A, sigma, X.0)` returns an $n \times 2$ matrix from the bivariate VAR(1) recursion

$$\vec{X}_i = \vec{X}_{i-1} A + \vec{\varepsilon}_i, \qquad \vec{\varepsilon}_i = (\varepsilon_{i,1}, \varepsilon_{i,2}),\quad \varepsilon_{i,j} \stackrel{\text{indep}}{\sim} \mathcal{N}(0, \sigma_j^2),\ 1 \le i \le n$$

initialized with $\vec{X}_0$ (length 2). $A$ is a $2 \times 2$ matrix of autoregression coefficients with all eigenvalues in $(-1, 1)$ for stationarity.

## R Implementation

```r
Part2j <- function(n, A, sigma, X.0) {
  noise_mat   <- cbind(rnorm(n, 0, sigma[1]), rnorm(n, 0, sigma[2]))
  observations <- matrix(0, nrow = n, ncol = 2)
  prev_row    <- X.0
  for (i in seq_len(n)) {
    prev_row          <- as.vector(prev_row %*% A) + noise_mat[i, ]
    observations[i, ] <- prev_row
  }
  return(observations)
}
```

```r
set.seed(1)

n_steps     <- 1e6
A_coef      <- matrix(c(
  0.32, 0.27,
  0.27, 0.32
), nrow = 2, ncol = 2, byrow = TRUE)
noise_sigma <- c(2.2, 2.98)
X_init      <- c(-3.14, -1.61)

cat("n           =", n_steps,       "\n")
cat("A:\n");          print(A_coef)
cat("eigen(A):\n");   print(eigen(A_coef)$values)
cat("sigma       =", noise_sigma,   "\n")
cat("X.0         =", X_init,        "\n")
```

## Computing the Results

```r
X <- Part2j(n = n_steps, A = A_coef, sigma = noise_sigma, X.0 = X_init)

sd_col_1     <- sd(X[, 1])
sd_col_2     <- sd(X[, 2])
cor_cols     <- cor(X[, 1], X[, 2])

cat("dim(X)                =", dim(X),  "\n")
cat("a. sd(X[,1])          =", sd_col_1, "\n")
cat("b. sd(X[,2])          =", sd_col_2, "\n")
cat("c. cor(X[,1], X[,2])  =", cor_cols, "\n")
```
