## Overview

`Part2h(n, pi.0, P)` returns the marginal distribution of $X_n$ from a finite-state Markov chain on $\{1, \ldots, k\}$:

$$\Pr(X_n = x) = \big(\pi_0\, P^{\,n}\big)_x, \qquad x \in \{1, \ldots, k\}$$

obtained by iterating $d \leftarrow d\, P$ a total of $n$ times starting from $d = \pi_0$.

## R Implementation

```r
Part2h <- function(n, pi.0, P) {
  d <- pi.0
  for (step_index in seq_len(n)) {
    d <- d %*% P
  }
  return(as.vector(d))
}
```

```r
n_step         <- 9
initial_dist   <- c(0.06, 0.79, 0.15)
transition_mat <- matrix(c(
  0.27, 0.21, 0.52,
  0.11, 0.11, 0.78,
  0.11, 0.72, 0.17
), nrow = 3, ncol = 3, byrow = TRUE)

cat("n              =", n_step,                 "\n")
cat("k (states)     =", length(initial_dist),   "\n")
cat("pi.0:\n");          print(initial_dist)
cat("rowSums(pi.0)  =", sum(initial_dist),      "\n")
cat("P:\n");             print(transition_mat)
cat("rowSums(P)     =", rowSums(transition_mat),"\n")
```

## Computing the Results

```r
marginal_at_n <- Part2h(n = n_step, pi.0 = initial_dist, P = transition_mat)

cat("Marginal P(X_n = x) for n =", n_step, ":\n")
cat("a. P(X_n = 1) =", marginal_at_n[1], "\n")
cat("b. P(X_n = 2) =", marginal_at_n[2], "\n")
cat("c. P(X_n = 3) =", marginal_at_n[3], "\n")
cat("sum           =", sum(marginal_at_n),  "\n")
```
