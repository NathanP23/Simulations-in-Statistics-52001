## Overview

`Part4c(P, alpha, N, x0, seed)` simulates a discrete Metropolis-Hastings Markov chain and estimates its stationary distribution from the empirical state frequencies.

At step $i$, the current state is $x = X_{i-1}$. A proposal $y$ is drawn from row $x$ of the proposal matrix $P$. The proposal is accepted with probability $\alpha(x,y)$, stored in the matrix `alpha`.

If the proposal is accepted, $X_i = y$; otherwise $X_i = x$.

**Construction:**

1. Choose an initial state `x0`.
2. Simulate `N` Metropolis-Hastings transitions using proposal matrix `P` and acceptance matrix `alpha`.
3. Store the full simulated process in `X`.
4. Compute `limit.pi <- table(X) / length(X)` as an approximation of the stationary distribution.


## R Implementation


```r
Part4c <- function(P, alpha, N, x0 = 1, seed = 1) {
  if (!is.null(seed)) set.seed(seed)

  n.states <- nrow(P)
  X <- numeric(N + 1)
  X[1] <- x0

  for (i in seq_len(N)) {
    current <- X[i]
    proposal <- sample(seq_len(n.states), size = 1, prob = P[current, ])

    if (runif(1) < alpha[current, proposal]) {
      X[i + 1] <- proposal
    } else {
      X[i + 1] <- current
    }
  }

  limit.pi <- table(factor(X, levels = seq_len(n.states))) / length(X)

  return(list(
    X = X,
    limit.pi = limit.pi
  ))
}
```

```r
P <- matrix(c(
  0.625, 0.150, 0.225,
  0.175, 0.325, 0.500,
  0.550, 0.150, 0.300
), nrow = 3, ncol = 3, byrow = TRUE)

alpha <- matrix(c(
  1.0000, 1.0000, 1.0000,
  0.7143, 1.0000, 1.0000,
  0.0705, 0.6897, 1.0000
), nrow = 3, ncol = 3, byrow = TRUE)

N <- 1e5
x0 <- 1
seed <- 1

cat("P:\n"); print(P)
cat("alpha:\n"); print(alpha)
cat("N    =", N, "\n")
cat("x0   =", x0, "\n")
cat("seed =", seed, "\n")
```

## Computing the Results


```r
mh_results <- Part4c(
  P = P,
  alpha = alpha,
  N = N,
  x0 = x0,
  seed = seed
)

X <- mh_results$X
limit.pi <- mh_results$limit.pi

cat("limit.pi:
"); print(limit.pi)
cat("
")
cat("a. limit.pi[1] =", limit.pi[1], "
")
cat("b. limit.pi[2] =", limit.pi[2], "
")
cat("c. limit.pi[3] =", limit.pi[3], "
")
```
