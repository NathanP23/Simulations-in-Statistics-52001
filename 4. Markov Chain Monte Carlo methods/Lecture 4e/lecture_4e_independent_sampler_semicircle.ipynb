## Overview

`Part4e(lower, upper, N, x0, seed)` simulates an independent-sampler Metropolis-Hastings chain targeting a density proportional to

$$\pi(x) \propto \sqrt{1 - x^2}, \qquad lower < x < upper.$$

The proposal is independent of the current state:

$$Y_i \sim U(lower, upper).$$

Since the proposal density is Uniform on the interval, the proposal ratio cancels out, so the acceptance probability is

$$\alpha(x, y) = \min\left\{\frac{\sqrt{1-y^2}}{\sqrt{1-x^2}}, 1\right\}.$$

**Construction:**

1. Start from `x0 = 0`.
2. Generate an independent proposal from `Uniform(lower, upper)`.
3. Accept the proposal using the target-density ratio above.
4. Store the simulated path `X_1, ..., X_N` in `X`.


## R Implementation


```r
Part4e <- function(lower, upper, N, x0 = 0, seed = 1) {
  if (!is.null(seed)) set.seed(seed)

  X <- numeric(N)
  current <- x0

  for (i in seq_len(N)) {
    proposal <- runif(1, min = lower, max = upper)
    alpha <- min(sqrt(1 - proposal^2) / sqrt(1 - current^2), 1)

    if (runif(1) < alpha) {
      current <- proposal
    }

    X[i] <- current
  }

  return(X)
}
```

```r
lower <- -0.72
upper <- 0.84
N <- 1e6
x0 <- 0
seed <- 1

cat("lower =", lower, "
")
cat("upper =", upper, "
")
cat("N     =", N, "
")
cat("x0    =", x0, "
")
cat("seed  =", seed, "
")
```

## Computing the Results


```r
X <- Part4e(
  lower = lower,
  upper = upper,
  N = N,
  x0 = x0,
  seed = seed
)

mean_changed <- mean(diff(X) != 0)
mean_sqrt <- mean(sqrt(1 - X^2))

cat("a. mean(diff(X) != 0)  =", mean_changed, "
")
cat("b. mean(sqrt(1-X^2))  =", mean_sqrt, "
")
```
