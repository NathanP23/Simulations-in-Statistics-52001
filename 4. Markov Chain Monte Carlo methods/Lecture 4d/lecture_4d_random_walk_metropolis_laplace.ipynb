## Overview

`Part4d(lam, a, N, x0, seed)` simulates a random-walk Metropolis-Hastings chain targeting the density proportional to

$$\pi(x) = \frac{\lambda}{2} e^{-\lambda |x|}, \qquad -\infty < x < \infty.$$

The proposal has the form

$$Y_i = X_{i-1} + \epsilon_i, \qquad \epsilon_i \sim U(-a, a).$$

Because the proposal distribution is symmetric, the acceptance probability is

$$\alpha(x, y) = \min\left\{\frac{\pi(y)}{\pi(x)}, 1\right\}
= \min\left\{e^{-\lambda(|y| - |x|)}, 1\right\}.$$

**Construction:**

1. Start from `x0 = 0`.
2. Generate a proposal by adding a Uniform increment on `[-a, a]` to the current state.
3. Accept the proposal with probability `exp(-lam * (abs(proposal) - abs(current)))`, capped at 1.
4. Store the simulated path `X_1, ..., X_N` in `X`.


## R Implementation


```r
Part4d <- function(lam, a, N, x0 = 0, seed = 1) {
  if (!is.null(seed)) set.seed(seed)

  X <- numeric(N)
  current <- x0

  for (i in seq_len(N)) {
    proposal <- current + runif(1, min = -a, max = a)
    alpha <- min(exp(-lam * (abs(proposal) - abs(current))), 1)

    if (runif(1) < alpha) {
      current <- proposal
    }

    X[i] <- current
  }

  return(X)
}
```

```r
lam <- 1.84
a <- 1.74
N <- 1e6
x0 <- 0
seed <- 1

cat("lam  =", lam, "
")
cat("a    =", a, "
")
cat("N    =", N, "
")
cat("x0   =", x0, "
")
cat("seed =", seed, "
")
```

## Computing the Results


```r
X <- Part4d(
  lam = lam,
  a = a,
  N = N,
  x0 = x0,
  seed = seed
)

mean_changed <- mean(diff(X) != 0)
mean_exp_abs <- mean(exp(-abs(X)))

cat("a. mean(diff(X) != 0) =", mean_changed, "
")
cat("b. mean(exp(-abs(X))) =", mean_exp_abs, "
")
```
