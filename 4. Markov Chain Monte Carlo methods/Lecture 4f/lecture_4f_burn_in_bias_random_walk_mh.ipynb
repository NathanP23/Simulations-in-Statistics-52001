## Overview

`Part4f(lam, a, N, n.sim, x0, seed)` produces many independent copies of

$$Z_N^{MH} = \frac{1}{N}\sum_{i=1}^{N} f(X_i), \qquad f(x)=e^{-|x|},$$

where $X_1, \ldots, X_N$ are generated by a random-walk Metropolis-Hastings chain targeting the Laplace density

$$\pi(x)=\frac{\lambda}{2}e^{-\lambda|x|}.$$

The proposal is

$$Y_i = X_{i-1} + \epsilon_i, \qquad \epsilon_i \sim U(-a,a),$$

and the symmetric proposal gives acceptance probability

$$\alpha(x,y)=\min\left\{e^{-\lambda(|y|-|x|)},1\right\}.$$

The provided `Part4f.R` file in this folder uses a different target distribution and Normal proposals, so the notebook uses the same vectorized many-chain idea but implements the exercise-specific model directly.


## R Implementation


```r
Part4f <- function(lam, a, N, n.sim, x0 = 0, seed = 1) {
  if (!is.null(seed)) set.seed(seed)

  X <- rep(x0, n.sim)
  S <- numeric(n.sim)

  for (i in seq_len(N)) {
    proposal <- X + runif(n.sim, min = -a, max = a)
    alpha <- pmin(exp(-lam * (abs(proposal) - abs(X))), 1)
    accept <- runif(n.sim) < alpha
    X[accept] <- proposal[accept]
    S <- S + exp(-abs(X))
  }

  Z.MH <- S / N
  target.expectation <- lam / (1 + lam)

  return(list(
    Z.MH = Z.MH,
    bias = mean(Z.MH) - target.expectation,
    sd = sd(Z.MH),
    target.expectation = target.expectation
  ))
}
```

```r
lam <- 0.0179
a <- 1.81
N <- 50
n.sim <- 1e5
x0 <- 0
seed <- 1

cat("lam   =", lam, "
")
cat("a     =", a, "
")
cat("N     =", N, "
")
cat("n.sim =", n.sim, "
")
cat("x0    =", x0, "
")
cat("seed  =", seed, "
")
```

## Computing the Results


```r
simulation_results <- Part4f(
  lam = lam,
  a = a,
  N = N,
  n.sim = n.sim,
  x0 = x0,
  seed = seed
)

Z.MH <- simulation_results$Z.MH
bias <- simulation_results$bias
sd.Z <- simulation_results$sd

cat("target expectation =", simulation_results$target.expectation, "
")
cat("mean(Z.MH)         =", mean(Z.MH), "
")
cat("
")
cat("a. bias            =", bias, "
")
cat("b. sd              =", sd.Z, "
")
```
