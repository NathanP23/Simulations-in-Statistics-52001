## Overview

`Part4b(target.pi, x0, P, Y, U)` reconstructs the states $X_0, X_1, \ldots, X_N$ of a discrete Metropolis-Hastings process.

At step $i$, the current state is $x = X_{i-1}$ and the proposal is $y = Y_i$. The proposal is accepted with probability

$$\alpha(x, y) = \min\left\{\frac{\pi_y P_{y,x}}{\pi_x P_{x,y}}, 1\right\}.$$

If $U_i < \alpha(x, y)$, then $X_i = y$; otherwise $X_i = x$.

**Construction:**

1. Start the chain at `x0`.
2. For each proposal `Y[i]`, compute the Metropolis-Hastings acceptance probability from the current state.
3. Accept the proposal when `U[i] < alpha`; otherwise repeat the previous state.
4. Return the full path, the final state `X.N`, and the number of transitions where the state changes.


## R Implementation


```r
Part4b <- function(target.pi, x0, P, Y, U) {
  N <- length(Y)
  X <- numeric(N + 1)
  X[1] <- x0

  for (i in seq_len(N)) {
    x <- X[i]
    y <- Y[i]
    alpha <- min(target.pi[y] * P[y, x] / (target.pi[x] * P[x, y]), 1)

    if (U[i] < alpha) {
      X[i + 1] <- y
    } else {
      X[i + 1] <- x
    }
  }

  return(list(
    X = X,
    X.N = X[N + 1],
    n.transitions = sum(X[-1] != X[-length(X)])
  ))
}
```

```r
target.pi <- c(0.15, 0.225, 0.125, 0.5)
x0 <- 2
P <- matrix(0.25, nrow = 4, ncol = 4)
N <- 20
Y <- c(2, 2, 2, 4, 3, 3, 2, 2, 3, 2, 4, 1, 3, 4, 3, 2, 4, 2, 1, 4)
U <- c(0.932, 0.229, 0.219, 0.701, 0.395, 0.49, 0.441, 0.62, 0.815, 0.396,
       0.43, 0.44, 0.735, 0.401, 0.869, 0.91, 0.516, 0.444, 0.532, 0.853)

cat("target.pi =", target.pi, "\n")
cat("x0        =", x0, "\n")
cat("P:\n"); print(P)
cat("N         =", N, "\n")
cat("Y         =", Y, "\n")
cat("U         =", U, "\n")
```

## Computing the Results


```r
mh_results <- Part4b(
  target.pi = target.pi,
  x0 = x0,
  P = P,
  Y = Y,
  U = U
)

X <- mh_results$X
X.N <- mh_results$X.N
n.transitions <- mh_results$n.transitions

cat("X =", X, "
")
cat("
")
cat("a. X_N           =", X.N, "
")
cat("b. transitions   =", n.transitions, "
")
```
