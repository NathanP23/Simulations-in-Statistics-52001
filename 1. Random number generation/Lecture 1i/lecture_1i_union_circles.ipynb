## Overview

We want to sample uniformly from the set $A$, defined as the **union of two circles** intersected with the unit square $[0,1] \times [0,1]$.

$$A = \Big(\{(x_1, x_2) : (x_1 - a_{1,1})^2 + (x_2 - a_{1,2})^2 \leq r^2\} \;\cup\; \{(x_1, x_2) : (x_1 - a_{2,1})^2 + (x_2 - a_{2,2})^2 \leq r^2\}\Big) \;\cap\; [0,1]^2$$

where:
- $a_1 = (a_{1,1},\; a_{1,2})$ — centre of the first circle
- $a_2 = (a_{2,1},\; a_{2,2})$ — centre of the second circle
- $r$ — common radius

**Proposal:** Uniform on $[0,1] \times [0,1]$.

**Accept** row $(X_{1i}, X_{2i})$ if the point falls inside **at least one** of the two circles:

$$\text{dist}(X_i, a_1) \leq r \quad \text{or} \quad \text{dist}(X_i, a_2) \leq r$$

## R Implementation

```r
euclidean_dist <- function(x1, x2, c1, c2) {
  sqrt((x1 - c1)^2 + (x2 - c2)^2)
}

Part1i <- function(X, a1, a2, r) {
  dist_to_circle1 <- euclidean_dist(X[, 1], X[, 2], a1[1], a1[2])
  dist_to_circle2 <- euclidean_dist(X[, 1], X[, 2], a2[1], a2[2])
  is_accepted <- (dist_to_circle1 <= r) | (dist_to_circle2 <= r)
  return(X[is_accepted, , drop = FALSE])
}
```

```r
X <- matrix(c(
  0.621, 0.294,
  0.971, 0.601,
  0.997, 0.564,
  0.021, 0.622,
  0.198, 0.897,
  0.640, 0.879
), ncol = 2, byrow = TRUE)

a1 <- c(0.4 , 0.04)
a2 <- c( 0.83 , 0.48)
r  <- 0.5

cat("a1 =", a1, "\n")
cat("a2 =", a2, "\n")
cat("r  =", r,  "\n")
cat("X:\n")
print(X)
```

## Computing the Results

```r
x_accepted          <- Part1i(X, a1, a2, r)
accepted_first_x1   <- x_accepted[1, 1]
accepted_first_x2   <- x_accepted[1, 2]
proportion_accepted <- nrow(x_accepted) / nrow(X)

cat("a. X.acc[1,1]            =", accepted_first_x1,   "\n")
cat("b. X.acc[1,2]            =", accepted_first_x2,   "\n")
cat("c. Proportion accepted   =", proportion_accepted,  "\n")
```

## Interpretation

- **Accept/reject rule:** Accept $(X_{1i}, X_{2i})$ if $\text{dist}(X_i, a_1) \leq r$ **or** $\text{dist}(X_i, a_2) \leq r$
- **X.acc[1,1] and X.acc[1,2]:** the first proposal that fell inside the union of the two circles
- **Proportion accepted:** estimates $\text{Area}(A) / 1 = \text{Area}(A)$ since the proposal is uniform on $[0,1]^2$

## Final Answer

Accept $(X_{1i}, X_{2i})$ if the point lies in at least one circle:

$$\|(X_i - a_1)\| \leq r \quad \text{or} \quad \|(X_i - a_2)\| \leq r$$

$$a_1 = (0.43,\; 0.7), \quad a_2 = (0.98,\; 0.58), \quad r = 0.48$$

$$\text{(a)}\ X.\text{acc}[1,1] \qquad \text{(b)}\ X.\text{acc}[1,2] \qquad \text{(c)}\ \frac{|\text{accepted rows}|}{|\text{proposal rows}|}$$
