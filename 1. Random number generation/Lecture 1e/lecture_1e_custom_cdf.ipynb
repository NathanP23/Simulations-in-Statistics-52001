## Overview

The target distribution has CDF:

$$F_X(x) = 1 - \frac{1}{x}, \quad x \geq 1 \qquad (F_X(x) = 0 \text{ otherwise})$$

This is a **Pareto distribution** with shape $\alpha = 1$ and scale $x_m = 1$.

Inverting $F_X$ to apply the **inverse transform method**:

$$U = 1 - \frac{1}{X} \implies X = \frac{1}{1 - U}, \quad U \sim U(0,1)$$

## R Implementation

```r
Part1e <- function(n, seed) {
  set.seed(seed)
  uniform_sample      <- runif(n)
  transformed_sample  <- 1 / (1 - uniform_sample)

  cdf_target <- function(x) ifelse(x >= 1, 1 - 1 / x, 0)
  ks_result  <- ks.test(transformed_sample, cdf_target)

  return(list(
    x_first       = transformed_sample[1],
    ks_p_value    = ks_result$p.value
  ))
}
```

```r
n_samples  <- 7549
seed_value <- 915

cat("Parameters:\n")
cat("  n    =", n_samples,  "\n")
cat("  seed =", seed_value, "\n")
```

## Computing the Results

```r
results <- Part1e(n = n_samples, seed = seed_value)

cat("a. X[1] (first transformed value) =", results$x_first,    "\n")
cat("b. KS test p-value                =", results$ks_p_value, "\n")
```

## Interpretation

- **X[1]** is the first simulated value: $1 / (1 - U_1)$ where $U_1$ is the first uniform draw
- **KS p-value** tests whether the transformed sample follows $F_X(x) = 1 - 1/x$ — a large p-value confirms the inverse transform is correct

## Final Answer

$$X_i = \frac{1}{1 - U_i}, \quad U_i \sim U(0,1)$$

$$\text{(a)}\ X[1] \qquad \text{(b)}\ \text{p-value} = \text{ks.test}\!\left(X,\ F_X(x) = 1 - \tfrac{1}{x}\right)$$
