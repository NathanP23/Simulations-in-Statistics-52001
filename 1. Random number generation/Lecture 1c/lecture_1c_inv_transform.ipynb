## Overview

The **inverse transform method** converts a $U(0,1)$ sample into a sample from any target distribution $F$ by applying the quantile function $F^{-1}$:

$$X = F^{-1}(U), \quad U \sim U(0,1) \implies X \sim F$$

Here we simulate from a $\chi^2(\text{df})$ distribution using `qchisq` as the inverse CDF, then apply the **Kolmogorov-Smirnov test** to both:

- the raw uniform sample (testing against $U(0,1)$)
- the transformed sample (testing against $\chi^2(\text{df})$)

## R Implementation

```r
Part1c <- function(n, df, seed) {
  set.seed(seed)
  uniform_sample     <- runif(n)
  chi_square_sample  <- qchisq(uniform_sample, df = df)

  ks_uniform         <- ks.test(uniform_sample,    "punif",   0, 1)
  ks_chi_square      <- ks.test(chi_square_sample, "pchisq",  df)

  return(list(
    p_value_uniform    = ks_uniform$p.value,
    p_value_chi_square = ks_chi_square$p.value
  ))
}
```

```r
n_samples        <- 7527
degrees_freedom  <- 10
seed_value       <- 116

cat("Parameters:\n")
cat("  n    =", n_samples,       "\n")
cat("  df   =", degrees_freedom, "\n")
cat("  seed =", seed_value,      "\n")
```

## Computing the p-values

```r
ks_results <- Part1c(n = n_samples, df = degrees_freedom, seed = seed_value)

cat("a. p-value (Uniform sample vs U(0,1))          =", ks_results$p_value_uniform,    "\n")
cat("b. p-value (Chi-square sample vs Chi-sq(df=4)) =", ks_results$p_value_chi_square, "\n")
```

## Interpretation

Both p-values should be large — neither test should reject its null hypothesis if the inverse transform was applied correctly:

- **p-value (a):** tests that `runif` output follows $U(0,1)$ — expected to be large
- **p-value (b):** tests that `qchisq(runif(n), df)` follows $\chi^2(\text{df})$ — expected to be large, confirming the inverse transform worked

## Final Answer

$$X = F^{-1}_{\chi^2(\text{df})}(U), \quad U \sim U(0,1)$$

$$\text{p-value}_a = \text{ks.test}(U,\ U(0,1)) \qquad \text{p-value}_b = \text{ks.test}(X,\ \chi^2(\text{df}))$$
