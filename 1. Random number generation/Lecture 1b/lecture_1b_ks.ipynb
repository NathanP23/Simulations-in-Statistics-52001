## Overview

The **Kolmogorov-Smirnov (KS) test** checks whether a sample comes from a specified distribution.
Here we use it to test whether R's built-in PRNG (`runif`) produces numbers that follow a $U(0,1)$ distribution.

The test statistic is:

$$D_n = \sup_x \left| F_n(x) - F(x) \right|$$

where $F_n$ is the empirical CDF of the sample and $F$ is the theoretical $U(0,1)$ CDF.
A large p-value gives no evidence against uniformity.

## R Implementation

```r
Part1b <- function(n, seed) {
  set.seed(seed)
  uniform_sample <- runif(n)
  ks_result      <- ks.test(uniform_sample, "punif", 0, 1)
  return(ks_result$p.value)
}
```

```r
n_samples  <- 7603
seed_value <- 462

cat("Parameters:\n")
cat("  n    =", n_samples,  "\n")
cat("  seed =", seed_value, "\n")
```

## Computing the p-value

```r
p_value <- Part1b(n = n_samples, seed = seed_value)
cat("p-value =", p_value, "\n")
```

## Interpretation

A p-value close to 1 means the sample is consistent with $U(0,1)$ — the KS test finds no evidence against uniformity.
A p-value below 0.05 would indicate the PRNG output deviates significantly from uniform.

## Final Answer

$$\text{p-value} = \text{ks.test}(\text{runif}(n),\ U(0,1))$$
