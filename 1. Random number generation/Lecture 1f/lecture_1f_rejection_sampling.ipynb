## Overview

We want to sample from a target density built via **rejection sampling**:

$$f_X(x) = \frac{1}{Z}\, p(x)\, g(x)$$

where:
- $g(x) = 1$ on $[0,1]$ — the **proposal** density ($U(0,1)$)
- $p(x) = 1 - 2|x - 0.5|$ — the **acceptance probability** function, $p(x) \in [0,1]$
- $Z = \int_0^1 p(x)\,g(x)\,dx$ — the normalising constant

Computing $Z$:

$$Z = \int_0^1 (1 - 2|x - 0.5|)\,dx = \int_0^{0.5} 2x\,dx + \int_{0.5}^1 (2 - 2x)\,dx = 0.25 + 0.25 = 0.5$$

So the target density is:

$$f_X(x) = 2(1 - 2|x - 0.5|), \quad x \in [0,1]$$

**Algorithm 1.19:** accept proposal $X_i$ if $U_i \leq p(X_i)$.

## R Implementation

```r
acceptance_prob <- function(x) 1 - 2 * abs(x - 0.5)
normalising_constant <- 0.5
density_target <- function(x) (1 / normalising_constant) * acceptance_prob(x)

Part1f <- function(x_proposals, u_acceptance) {
  is_accepted <- u_acceptance <= acceptance_prob(x_proposals)
  return(x_proposals[is_accepted])
}
```

```r
x_proposals   <- c(0.519, 0.251, 0.118, 0.372, 0.902, 0.458, 0.272)
u_acceptance  <- c(0.921, 0.231, 0.795, 0.143, 0.777, 0.212, 0.042)

cat("X =", x_proposals,  "\n")
cat("U =", u_acceptance, "\n")
```

## Computing the Results

```r
x_accepted            <- Part1f(x_proposals, u_acceptance)
accepted_first_value  <- x_accepted[1]
density_at_first      <- density_target(accepted_first_value)
proportion_accepted   <- length(x_accepted) / length(x_proposals)

cat("a. X.acc[1]                    =", accepted_first_value, "\n")
cat("b. f_X(X.acc[1])               =", density_at_first,    "\n")
cat("c. Proportion accepted         =", proportion_accepted,  "\n")
```

## Interpretation

- **X.acc[1]:** the first proposal that passed the acceptance test $U_i \leq p(X_i)$
- **$f_X$(X.acc[1]):** the target density evaluated at the first accepted value, $2(1 - 2|x - 0.5|)$
- **Proportion accepted:** converges to $Z = 0.5$ as the number of proposals grows — half the proposals are expected to be accepted on average

## Final Answer

Accept $X_i$ if $U_i \leq p(X_i) = 1 - 2|X_i - 0.5|$

$$f_X(x) = \frac{1}{Z}\,p(x) = 2(1 - 2|x - 0.5|), \quad Z = 0.5$$

$$\text{(a)}\ X.\text{acc}[1] \qquad \text{(b)}\ f_X(X.\text{acc}[1]) \qquad \text{(c)}\ \frac{|\text{accepted}|}{|\text{proposals}|}$$
