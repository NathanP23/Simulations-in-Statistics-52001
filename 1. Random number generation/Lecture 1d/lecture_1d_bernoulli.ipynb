## Overview

The **inverse transform method** for a $B(1, p)$ Bernoulli variable is:

$$X = \mathbf{1}[U < p], \quad U \sim U(0,1) \implies X \sim B(1, p)$$

Since the Bernoulli CDF jumps from $0$ to $p$ at $x=0$ and from $p$ to $1$ at $x=1$, the inverse transform reduces to a simple threshold comparison.

## R Implementation

```r
Part1d <- function(n, p, seed) {
  set.seed(seed)
  uniform_sample    <- runif(n)
  bernoulli_sample  <- as.integer(uniform_sample < p)

  return(list(
    x_first   = bernoulli_sample[1],
    x_average = mean(bernoulli_sample)
  ))
}
```

```r
n_samples    <- 7502
prob_success <- 0.685
seed_value   <- 838

cat("Parameters:\n")
cat("  n    =", n_samples,    "\n")
cat("  p    =", prob_success, "\n")
cat("  seed =", seed_value,   "\n")
```

## Computing the Results

```r
results <- Part1d(n = n_samples, p = prob_success, seed = seed_value)

cat("a. X[1] (first Bernoulli value) =", results$x_first,   "\n")
cat("b. mean(X)                      =", results$x_average, "\n")
```

## Interpretation

- **X[1]** is either 0 or 1, determined by whether the first uniform draw fell below $p$
- **mean(X)** is the sample proportion of 1s — by the Law of Large Numbers it should converge to $p = 0.379$ as $n$ grows

## Final Answer

$$X_i = \mathbf{1}[U_i < p], \quad U_i \sim U(0,1)$$

$$\text{(a)}\ X[1] \qquad \text{(b)}\ \bar{X} = \frac{1}{n}\sum_{i=1}^{n} X_i$$
