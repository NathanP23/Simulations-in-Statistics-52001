## Overview

The **Dirichlet distribution** with parameter $\alpha = (\alpha_1, \dots, \alpha_d)$ has joint density:

$$f(x_1, \dots, x_d) = \frac{\Gamma(\alpha_1 + \cdots + \alpha_d)}{\Gamma(\alpha_1)\,\Gamma(\alpha_2)\cdots\Gamma(\alpha_d)}\;x_1^{\alpha_1 - 1}\,x_2^{\alpha_2 - 1}\cdots x_d^{\alpha_d - 1}$$

for $x_1 + x_2 + \cdots + x_d = 1$, $x_j \geq 0$.

**Sampling via Gamma variates:** If $Y_j \sim \text{Gamma}(\alpha_j, 1)$ independently for $1 \leq j \leq d$, then:

$$X_j = \frac{Y_j}{Y_1 + Y_2 + \cdots + Y_d}, \quad 1 \leq j \leq d$$

follows a Dirichlet distribution with parameter $(\alpha_1, \alpha_2, \dots, \alpha_d)$.

## R Implementation

```r
Part1j <- function(n, alpha) {
  dimension <- length(alpha)
  gamma_matrix <- matrix(nrow = n, ncol = dimension)
  for (col_index in 1:dimension) {
    gamma_matrix[, col_index] <- rgamma(n, shape = alpha[col_index], rate = 1)
  }
  row_sums <- rowSums(gamma_matrix)
  return(gamma_matrix / row_sums)
}
```

```r
sample_size <- 10^6
alpha <- c(4.4, 1.9, 6.8, 4.5, 8.3, 1.1)

cat("n     =", sample_size, "\n")
cat("alpha =", alpha, "\n")

set.seed(NULL)
X <- Part1j(sample_size, alpha)
```

## Computing the Results

```r
mean_x1 <- mean(X[, 1])
sd_x1   <- sd(X[, 1])

cat("a. mean(X[,1]) =", mean_x1, "\n")
cat("b. sd(X[,1])   =", sd_x1,   "\n")
```

## Interpretation

For the Dirichlet distribution with parameter $\alpha = (\alpha_1, \dots, \alpha_d)$, the theoretical moments of $X_j$ are:

$$E[X_j] = \frac{\alpha_j}{\alpha_0}, \qquad \text{Var}(X_j) = \frac{\alpha_j(\alpha_0 - \alpha_j)}{\alpha_0^2(\alpha_0 + 1)}$$

where $\alpha_0 = \sum_{j=1}^d \alpha_j$.

For $X_1$ with $\alpha_1 = 8.4$ and $\alpha_0 = 8.4 + 7.9 + 2.8 + 7.8 = 26.9$:

$$E[X_1] = \frac{8.4}{26.9} \approx 0.3123, \qquad \text{SD}(X_1) = \sqrt{\frac{8.4 \times 18.5}{26.9^2 \times 27.9}} \approx 0.0877$$

The sample mean and SD should be close to these theoretical values with $n = 10^6$.

## Final Answer

Sample $Y_j \sim \text{Gamma}(\alpha_j, 1)$ independently, then normalise: $X_j = Y_j / \sum Y_k$.

$$\alpha = (8.4,\; 7.9,\; 2.8,\; 7.8), \quad n = 10^6$$

$$\text{(a)}\ \text{mean}(X[,1]) \approx 0.3123 \qquad \text{(b)}\ \text{sd}(X[,1]) \approx 0.0877$$
