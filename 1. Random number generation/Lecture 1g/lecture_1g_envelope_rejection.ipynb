## Overview

We want to sample from a target density using **envelope rejection sampling**:

$$\tilde{f}_X(x) = \frac{1}{Z_f}\,f(x) = \frac{1}{Z_f}\,|\cos(x)|\,e^{-\lambda x}, \quad x > 0$$

where:
- $\lambda = 1.8$
- $f(x) = |\cos(x)|\,e^{-\lambda x}$ — the unnormalised target
- $Z_f = \int_0^\infty f(x)\,dx$ — the normalising constant
- $g(x) = \lambda e^{-\lambda x}$ — the **proposal** density (Exponential($\lambda$))

**Finding the envelope constant $c$:**

$$c = \max_{x > 0} \frac{f(x)}{g(x)} = \max_{x > 0} \frac{|\cos(x)|\,e^{-\lambda x}}{\lambda\,e^{-\lambda x}} = \max_{x > 0} \frac{|\cos(x)|}{\lambda} = \frac{1}{\lambda}$$

since $\max_{x > 0} |\cos(x)| = 1$.

**Algorithm 1.22:** Accept proposal $X_i$ if $U_i \leq \dfrac{f(X_i)}{c \cdot g(X_i)} = |\cos(X_i)|$.

## R Implementation

```r
lambda_rate <- 1.8
envelope_constant <- 1 / lambda_rate

acceptance_ratio <- function(x) abs(cos(x))

Part1g <- function(X, U) {
  is_accepted <- U <= acceptance_ratio(X)
  return(X[is_accepted])
}
```

```r
X <- c(0.217, 0.248, 0.201, 0.123, 0.037, 0.191, 0.171, 0.049, 0.648)
U <- c(0.254, 0.884, 0.112, 0.951, 0.285, 0.76, 0.488, 0.527, 0.576)

cat("lambda =", lambda_rate, "\n")
cat("c      =", envelope_constant, "\n")
cat("X =", X, "\n")
cat("U =", U, "\n")
```

## Computing the Results

```r
x_accepted           <- Part1g(X, U)
accepted_first_value <- x_accepted[1]
proportion_accepted  <- length(x_accepted) / length(X)

cat("a. X.acc[1]              =", accepted_first_value, "\n")
cat("b. Proportion accepted   =", proportion_accepted,  "\n")
```

## Interpretation

- **Accept/reject rule:** Accept $X_i$ if $U_i \leq |\cos(X_i)|$
- **X.acc[1]:** the first proposal from the Exponential($\lambda$) distribution that passed the acceptance test
- **Proportion accepted:** the fraction of proposals accepted out of total proposals

## Final Answer

Accept $X_i$ if $U_i \leq |\cos(X_i)|$

$$\tilde{f}_X(x) = \frac{1}{Z_f}\,|\cos(x)|\,e^{-\lambda x}, \quad \lambda = 1.8, \quad c = \frac{1}{\lambda}$$

$$\text{(a)}\ X.\text{acc}[1] \qquad \text{(b)}\ \frac{|\text{accepted}|}{|\text{proposals}|}$$
