## Overview

We use the **ratio-of-uniforms** method to sample from the Exponential($\lambda$) distribution with density $f(x) = \lambda e^{-\lambda x}$.

The method states that if $(U, V)$ is sampled uniformly over the set:

$$A = \{(u, v) : 0 \leq u^2 \leq f(v/u)\}$$

then $V/U$ has density proportional to $f$.

**Rewriting the set** for $f(x) = \lambda e^{-\lambda x}$:

$$u^2 \leq \lambda e^{-\lambda v/u} \;\;\Longleftrightarrow\;\; 0 \leq v \leq g(u) = -\frac{u}{\lambda}\log\!\Big(\frac{u^2}{\lambda}\Big)$$

**Bounding rectangle:** The function $g(u)$ is positive on $(0, \sqrt{\lambda})$.

- $g'(u) = -\frac{1}{\lambda}\log(u^2/\lambda) - \frac{2}{\lambda} = 0 \;\Rightarrow\; \hat{u} = \sqrt{\lambda/e}$
- $g(\hat{u}) = \frac{2}{e\sqrt{\lambda}}$

So $A \subseteq [0, \sqrt{\lambda}] \times [0, 2/(e\sqrt{\lambda})]$.

**Input:** $X_0 \sim U(0, \sqrt{\lambda})$ and $X_1 \sim U(0, 2/(e\sqrt{\lambda}))$.

**Accept** if $X_1 \leq g(X_0)$. The accepted sample from Exp($\lambda$) is $X_1 / X_0$.

## R Implementation

```r
g_bound <- function(u, lam) {
  -(u / lam) * log(u^2 / lam)
}

Part1k <- function(X, lam) {
  x0 <- X[, 1]
  x1 <- X[, 2]
  upper_bound <- g_bound(x0, lam)
  is_accepted <- (x1 >= 0) & (x1 <= upper_bound)
  return(X[is_accepted, , drop = FALSE])
}
```

```r
X <- matrix(c(
  1.441, 0.143,
  1.501, 0.003,
  0.772, 0.051,
  0.663, 0.091,
  0.681, 0.295,
  0.381, 0.058
), ncol = 2, byrow = TRUE)

colnames(X) <- c("X0", "X1")

lam <- 4.6

cat("lam =", lam, "\n")
cat("sqrt(lam)         =", sqrt(lam), "\n")
cat("2/(e*sqrt(lam))   =", 2 / (exp(1) * sqrt(lam)), "\n")
cat("X:\n")
print(X)
```

## Computing the Results

```r
x_accepted          <- Part1k(X, lam)
exponential_samples <- x_accepted[, 2] / x_accepted[, 1]
first_sample        <- exponential_samples[1]
proportion_accepted <- nrow(x_accepted) / nrow(X)

cat("a. First sample (X1/X0)  =", first_sample,        "\n")
cat("b. Proportion accepted   =", proportion_accepted,  "\n")
```

## Interpretation

- **Accept/reject rule:** Accept row $(X_0, X_1)$ if $0 \leq X_1 \leq g(X_0) = -\frac{X_0}{\lambda}\log(X_0^2/\lambda)$
- **Exponential sample:** For each accepted row, $X_1/X_0 \sim \text{Exp}(\lambda)$
- **Proportion accepted:** the fraction of proposal pairs that fell inside the set $A$

## Final Answer

Accept $(X_0, X_1)$ if $X_1 \leq g(X_0) = -\frac{X_0}{\lambda}\log(X_0^2/\lambda)$. The Exp($\lambda$) sample is $X_1/X_0$.

$$\lambda = 3.4, \quad A \subseteq [0, \sqrt{3.4}] \times [0,\; 2/(e\sqrt{3.4})]$$

$$\text{(a) first accepted } X_1/X_0 \qquad \text{(b) proportion accepted}$$
