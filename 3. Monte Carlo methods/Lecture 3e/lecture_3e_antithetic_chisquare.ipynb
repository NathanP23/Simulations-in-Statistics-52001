## Overview

`Part3e(U, df)` uses antithetic variables to approximate $E(f(X)) = E(X^2)$ for a Chi-square random variable with `df` degrees of freedom.

**Construction:**

1. Transform each uniform value using `qchisq(U, df)` to get $X_{i1}$.
2. Transform the antithetic values using `qchisq(1 - U, df)` to get $X_{i2}$.
3. Compute $f(x) = x^2$ for both sequences.
4. Compute the antithetic Monte Carlo average.
5. Compute the empirical correlations between the two generated sequences and between the two transformed function-value sequences.

## R Implementation

```r
Part3e <- function(U, df) {
  X1 <- qchisq(U, df = df)
  X2 <- qchisq(1 - U, df = df)
  f1 <- X1^2
  f2 <- X2^2
  Z.av <- mean(f1 + f2) / 2
  cor.X <- cor(X1, X2)
  cor.f <- cor(f1, f2)
  return(c(Z.av = Z.av, cor.X = cor.X, cor.f = cor.f))
}
```

```r
U <- c(0.514, 0.756, 0.641, 0.047, 0.807)
degrees_freedom <- 5

cat("U  =", U, "\n")
cat("df =", degrees_freedom, "\n")
```

## Computing the Results

```r
antithetic_results <- Part3e(
  U  = U,
  df = degrees_freedom
)

cat("a. Z_AV  =", antithetic_results["Z.av"], "\n")
cat("b. cor X =", antithetic_results["cor.X"], "\n")
cat("c. cor f =", antithetic_results["cor.f"], "\n")
```
