## Overview

`Part3b(N, mu, Sigma)` simulates independent copies from a multivariate Normal distribution with mean vector `mu` and covariance matrix `Sigma`, applies the function $g(\vec{x}) = \cos^2(\|\vec{x}\|)$, and returns the Monte Carlo average.

**Construction:**

1. Draw $\vec{X}_1, \ldots, \vec{X}_N \stackrel{\text{iid}}{\sim} \mathcal{N}(\vec{\mu}, \Sigma)$ using `MASS::mvrnorm`.
2. Compute the Euclidean norm $\|\vec{X}_i\|$ for each simulated vector.
3. Apply $g(\vec{X}_i) = \cos^2(\|\vec{X}_i\|)$.
4. Return the Monte Carlo average $N^{-1}\sum_{i=1}^N g(\vec{X}_i)$.

## R Implementation

```r
suppressMessages(library(MASS))

Part3b <- function(N, mu, Sigma) {
  X <- mvrnorm(N, mu = mu, Sigma = Sigma)
  norms <- sqrt(rowSums(X^2))
  return(mean(cos(norms)^2))
}
```

```r
set.seed(1)

n_replicates <- 1e5
mu_vec <- c(-1.48, -0.87, 1.49, 0.75, -1.45)
Sigma_mat <- matrix(-0.1, nrow = 5, ncol = 5)
diag(Sigma_mat) <- 0.7

cat("N     =", n_replicates, "\n")
cat("mu    =", mu_vec,       "\n")
cat("Sigma:\n"); print(Sigma_mat)
```

## Computing the Results

```r
integral_value <- Part3b(
  N     = n_replicates,
  mu    = mu_vec,
  Sigma = Sigma_mat
)

cat("a. integral =", integral_value, "\n")
```
