## Overview

`Part3i(M, N)` computes the empirical correlation between `X` and `Y` in a data frame `M`, then estimates the standard deviation of this estimator using the non-parametric bootstrap.

**Construction:**

1. Compute the empirical correlation using `with(M, cor(X, Y))`.
2. For each bootstrap replication, sample rows from `M` with replacement.
3. Compute the empirical correlation in the simulated data frame.
4. Return the original empirical correlation and the standard deviation of the bootstrap correlations.

## R Implementation

```r
Part3i <- function(M, N) {
  n <- nrow(M)
  rho_hat <- with(M, cor(X, Y))
  rho_boot <- replicate(N, {
    M_boot <- M[sample(seq_len(n), n, replace = TRUE), ]
    with(M_boot, cor(X, Y))
  })
  return(c(correlation = rho_hat, sd = sd(rho_boot)))
}
```

```r
M <- data.frame(
  X = c(1.95, 1.07, 0.43, 2.07, 0.43, -0.59, -0.56, 1.31, -0.95, 0.71, 0.48, -1.15, 0.54, 0.21, 0.39),
  Y = c(3.26, 2.62, 0.52, 2.69, 0.92, -2.35, -0.64, -0.41, 0.85, 0.82, 1.66, -0.21, -0.31, 0.57, 1.09)
)
n_replicates <- 1e4

print(M)
cat("reps =", n_replicates, "\n")
```

## Computing the Results

```r
correlation_results <- Part3i(
  M = M,
  N = n_replicates
)

cat("a. correlation =", correlation_results["correlation"], "\n")
cat("b. sd          =", correlation_results["sd"], "\n")
```
