# Lecture 3g: Coverage of a t-test Confidence Interval
**Simulations in Statistics (52001)**

## Overview

`Part3g(n, rate, conf.level, N)` estimates the actual coverage probability of the confidence interval returned by `t.test` for samples from an Exponential distribution.

**Construction:**

1. The true expectation is $E(X) = 1 / rate$.
2. Simulate `N` independent samples of size `n` from `rexp(n, rate = rate)`.
3. For each sample, compute `t.test(sample, conf.level = conf.level)$conf.int`.
4. Check whether the true expectation is inside the interval.
5. Return the relative frequency of intervals that cover the true expectation.

## R Implementation

```r
Part3g <- function(n, rate, conf.level, N) {
  expectation <- 1 / rate
  covered <- replicate(N, {
    X <- rexp(n, rate = rate)
    interval <- t.test(X, conf.level = conf.level)$conf.int
    interval[1] <= expectation && expectation <= interval[2]
  })
  return(mean(covered))
}
```

```r
set.seed(1)

sample_size <- 10
rate_parameter <- 1.83
confidence_level <- 0.965
n_replicates <- 1e5

cat("n          =", sample_size, "\n")
cat("rate       =", rate_parameter, "\n")
cat("conf.level =", confidence_level, "\n")
cat("reps       =", n_replicates, "\n")
```

## Computing the Results

```r
coverage_probability <- Part3g(
  n          = sample_size,
  rate       = rate_parameter,
  conf.level = confidence_level,
  N          = n_replicates
)

cat("a. coverage probability =", coverage_probability, "\n")
```
