## Overview

`Part3d(X, Y, a)` compares the exact standard Normal tail probability, the standard Monte Carlo estimate based on a Normal sample `X`, and the importance sampling estimate based on a shifted Exponential sample `Y`.

**Construction:**

1. Compute the exact probability $P(X \ge a)$ using `1 - pnorm(a)`.
2. Compute the standard Monte Carlo estimate using `mean(X >= a)`.
3. For the shifted Exponential density $g(y) = a exp(-a(y-a))$ for $y \ge a$, compute the likelihood ratio $\phi(y) / g(y)$.
4. Return the three requested values.

## R Implementation

```r
Part3d <- function(X, Y, a) {
  exact_probability <- 1 - pnorm(a)
  standard_mc <- mean(X >= a)
  g_y <- a * exp(-a * (Y - a))
  importance_sampling <- mean(dnorm(Y) / g_y)
  return(c(
    exact = exact_probability,
    standard = standard_mc,
    importance = importance_sampling
  ))
}
```

```r
X <- c(
  -0.94, 1.18, 1.09, 2, 0.25, 1.9, 0.69, -1.99, -1.3, 1.02, -0.04, -0.12, 0.7, -0.54, 0.77, 0.63, 0.45, 0.22, -0.85, -0.43, 2.52, 0.56, 0.14, -0.05, 1.3)
Y <- c(2.86, 2.27, 2.19, 1.96, 1.92)
threshold_a <- 1.79

cat("X =", X, "\n")
cat("Y =", Y, "\n")
cat("a =", threshold_a, "\n")
```

## Computing the Results

```r
tail_results <- Part3d(
  X = X,
  Y = Y,
  a = threshold_a
)

cat("a. exact probability =", tail_results["exact"], "\n")
cat("b. standard MC       =", tail_results["standard"], "\n")
cat("c. importance        =", tail_results["importance"], "\n")
```
