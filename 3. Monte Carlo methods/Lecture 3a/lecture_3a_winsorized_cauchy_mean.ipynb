## Overview

`Part3a(n, location, alpha)` simulates one sample from a Cauchy distribution with location parameter `location`, winsorizes the sample using empirical percentiles `alpha` and `1 - alpha`, and returns the sample average of the winsorized values.

**Construction:**

1. Draw $X_1, \ldots, X_n \stackrel{\text{iid}}{\sim} \operatorname{Cauchy}(\text{location}, 1)$ using `rcauchy(n, location)`.
2. Compute the empirical percentiles $x_\alpha$ and $x_{1-\alpha}$ using `quantile(X, c(alpha, 1 - alpha))`.
3. Replace all values below $x_\alpha$ by $x_\alpha$ and all values above $x_{1-\alpha}$ by $x_{1-\alpha}$.
4. Return the sample average of the winsorized sample.
5. Repeat this procedure $10^5$ times and compute `mean(X.bar)`.

## R Implementation

```r
Part3a <- function(n, location, alpha) {
  X <- rcauchy(n, location)
  cutoffs <- quantile(X, c(alpha, 1 - alpha))
  X[X < cutoffs[1]] <- cutoffs[1]
  X[X > cutoffs[2]] <- cutoffs[2]
  return(mean(X))
}
```

```r
set.seed(1)

sample_size        <- 220
location_parameter <- 6.49
winsor_fraction    <- 0.058
n_replicates       <- 1e5

cat("n        =", sample_size,        "\n")
cat("location =", location_parameter, "\n")
cat("alpha    =", winsor_fraction,    "\n")
cat("reps     =", n_replicates,       "\n")
```

## Computing the Results

```r
X.bar <- replicate(
  n_replicates,
  Part3a(
    n        = sample_size,
    location = location_parameter,
    alpha    = winsor_fraction
  )
)

mean_X_bar <- mean(X.bar)

cat("length(X.bar)   =", length(X.bar), "\n")
cat("a. mean(X.bar) =", mean_X_bar,     "\n")
```
