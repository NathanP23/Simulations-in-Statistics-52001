## Overview

`Part3j(X, Y, m, N)` fits a logistic regression model to Binomial response data and estimates the standard deviation of the slope estimator using the parametric bootstrap.

**Construction:**

1. Fit the logistic regression model using `glm(cbind(Y, m - Y) ~ X, family = binomial)`.
2. Extract the estimated intercept and slope using `coef`.
3. Compute fitted probabilities from the fitted linear predictor.
4. Simulate `N` new Binomial response vectors using `rbinom(n, m, prob.hat)`.
5. Refit the logistic regression model to each simulated response vector and compute the bootstrap standard deviation of the slope estimates.

## R Implementation

```r
Part3j <- function(X, Y, m, N) {
  n <- length(X)
  fit <- glm(cbind(Y, m - Y) ~ X, family = binomial)
  coefficients <- coef(fit)
  inter.hat <- unname(coefficients[1])
  slope.hat <- unname(coefficients[2])
  pred.hat <- inter.hat + slope.hat * X
  prob.hat <- 1 / (1 + exp(-pred.hat))
  slope.boot <- replicate(N, {
    Y.boot <- rbinom(n, m, prob.hat)
    fit.boot <- glm(cbind(Y.boot, m - Y.boot) ~ X, family = binomial)
    unname(coef(fit.boot)[2])
  })
  return(c(intercept = inter.hat, slope = slope.hat, sd = sd(slope.boot)))
}
```

```r
sample_size <- 8
m_trials <- 17
X <- 1:8
Y <- c(12, 6, 10, 13, 11, 13, 15, 10)
n_replicates <- 1e4

cat("n    =", sample_size, "\n")
cat("m    =", m_trials, "\n")
cat("X    =", X, "\n")
cat("Y    =", Y, "\n")
cat("reps =", n_replicates, "\n")
```

## Computing the Results

```r
logistic_results <- Part3j(
  X = X,
  Y = Y,
  m = m_trials,
  N = n_replicates
)

cat("a. intercept =", logistic_results["intercept"], "\n")
cat("b. slope     =", logistic_results["slope"], "\n")
cat("c. sd        =", logistic_results["sd"], "\n")
```
