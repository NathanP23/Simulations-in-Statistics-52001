## Overview

`Part3c(N, Z, var.hat, alpha)` computes a normal-approximation confidence interval for the exact value of an integral estimated by Monte Carlo.

**Construction:**

1. Use the Monte Carlo estimate `Z`.
2. Use `var.hat` as the estimated variance of the simulated function values.
3. Compute the standard error `sqrt(var.hat / N)`.
4. Use the normal quantile `qnorm(1 - alpha / 2)` for a confidence level $1 - \alpha$.
5. Return the lower and upper endpoints of the confidence interval.

## R Implementation

```r
Part3c <- function(N, Z, var.hat, alpha) {
  z_alpha <- qnorm(1 - alpha / 2)
  se <- sqrt(var.hat / N)
  lower <- Z - z_alpha * se
  upper <- Z + z_alpha * se
  return(c(lower = lower, upper = upper))
}
```

```r
sample_size <- 1442
mc_estimate <- -4.29
var_hat <- 4.02
alpha_level <- 0.028

cat("N       =", sample_size, "\n")
cat("Z       =", mc_estimate, "\n")
cat("var.hat =", var_hat,    "\n")
cat("alpha   =", alpha_level, "\n")
```

## Computing the Results

```r
confidence_interval <- Part3c(
  N       = sample_size,
  Z       = mc_estimate,
  var.hat = var_hat,
  alpha   = alpha_level
)

cat("a. lower =", confidence_interval["lower"], "\n")
cat("b. upper =", confidence_interval["upper"], "\n")
```
