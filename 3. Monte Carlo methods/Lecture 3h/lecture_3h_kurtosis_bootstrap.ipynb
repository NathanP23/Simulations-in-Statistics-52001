## Overview

`Part3h(X, N)` computes the sample kurtosis estimator and estimates its standard deviation using both non-parametric and parametric bootstrap.

**Construction:**

1. Compute $\hat{\beta}$ using the ratio of the fourth centered sample moment to the squared second centered sample moment.
2. For the non-parametric bootstrap, repeatedly sample `n` values from `X` with replacement and evaluate $\hat{\beta}$.
3. For the parametric bootstrap, estimate the Exponential rate by `1 / mean(X)`, repeatedly sample from `rexp(n, rate = 1 / mean(X))`, and evaluate $\hat{\beta}$.
4. Return the original estimate and the two bootstrap standard deviations.

## R Implementation

```r
kurtosis_hat <- function(X) {
  centered <- X - mean(X)
  m2 <- mean(centered^2)
  m4 <- mean(centered^4)
  return(m4 / m2^2)
}

Part3h <- function(X, N) {
  n <- length(X)
  beta_hat <- kurtosis_hat(X)
  beta_np <- replicate(N, kurtosis_hat(sample(X, n, rep = TRUE)))
  beta_p <- replicate(N, kurtosis_hat(rexp(n, rate = 1 / mean(X))))
  return(c(
    beta = beta_hat,
    nonparametric = sd(beta_np),
    parametric = sd(beta_p)
  ))
}
```

```r
X <- c(0.24, 0.15, 0.78, 0.48, 0.08, 0.09, 2.87, 0.19, 0.26, 0.15, 0.21)
sample_size <- length(X)
n_replicates <- 1e5

cat("n    =", sample_size, "\n")
cat("X    =", X, "\n")
cat("reps =", n_replicates, "\n")
```

## Computing the Results

```r
bootstrap_results <- Part3h(
  X = X,
  N = n_replicates
)

cat("a. beta hat              =", bootstrap_results["beta"], "\n")
cat("b. sd non-parametric    =", bootstrap_results["nonparametric"], "\n")
cat("c. sd parametric        =", bootstrap_results["parametric"], "\n")
```
