## Overview

`Part3f(X, shape, rate, b)` computes the standard Monte Carlo estimate and the control-variate estimate for $E(f(X))$, where $f(x)=|x|^{2.5}$ and the control variate is $g(x)=x^2$.

**Construction:**

1. Compute the standard Monte Carlo estimate using `mean(abs(X)^2.5)`.
2. Use the known control-variate expectation $E(g(X)) = E(X^2) = 2r / \lambda^2$.
3. Compute the control-variate estimate `mean(f(X)) - b * mean(g(X) - E.g)`.
4. Return the standard and control-variate estimates.

## R Implementation

```r
Part3f <- function(X, shape, rate, b) {
  f_X <- abs(X)^2.5
  g_X <- X^2
  E_g <- 2 * shape / rate^2
  standard_mc <- mean(f_X)
  control_variate <- standard_mc - b * mean(g_X - E_g)
  return(c(standard = standard_mc, control = control_variate))
}
```

```r
X <- c(-0.29, 1.11, -2.12, -1.77, -0.97, 0.77, 1.62, -0.03, -0.46, -0.83)
shape_parameter <- 1.66
rate_parameter <- 1.28
slope_b <- 2.2

cat("X     =", X, "\n")
cat("shape =", shape_parameter, "\n")
cat("rate  =", rate_parameter, "\n")
cat("b     =", slope_b, "\n")
```

## Computing the Results

```r
control_results <- Part3f(
  X     = X,
  shape = shape_parameter,
  rate  = rate_parameter,
  b     = slope_b
)

cat("a. standard MC       =", control_results["standard"], "\n")
cat("b. control variate   =", control_results["control"], "\n")
```
